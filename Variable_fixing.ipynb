```python
import pandas as pd
import ast
pd.set_option('display.max_rows', 1000)
```

```python
hikes_df = pd.read_csv("hikes_df.csv", index_col = 0)
```

```python
hikes_df["rating_count"] = [int(i.strip("()").split(" ")[0]) for i in hikes_df["rating_count"]]
```

```python
hikes_df["rating"] = [float(i) for i in hikes_df["rating"]]
```

```python
a = []
for i in hikes_df["highpoint"]:
    try:
        a.append(int(str(i).split(":")[1].split(" ")[1]))
    except:
        a.append(None)
hikes_df["highpoint"] = a
```

```python
a = []
for i in hikes_df["gain"]:
    try:
        a.append(int(str(i).split(":")[1].split(" ")[1]))
    except:
        a.append(None)
hikes_df["gain"] = a
```

```python
a = []
for i in hikes_df["Length"]:
    try:
        if i.split("miles")[1].strip() == ", roundtrip":
            a.append(float(i.split("miles")[0]))
        elif i.split("miles")[1].strip() == "of trails":
            a.append(2* float(i.split("miles")[0]))
        elif i.split("miles")[1].strip() == ", one-way":
            a.append(2* float(i.split("miles")[0]))
    except:
        a.append(None)
hikes_df["Length"] = a
```

```python
hikes_df.rename(columns = {"Length":"length_miles", "gain": "gain_ft", "highpoint":"highpoint_ft"})
```
