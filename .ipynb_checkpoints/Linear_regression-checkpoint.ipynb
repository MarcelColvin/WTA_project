```python
import statsmodels.api as sm
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
hikes_df = pd.read_csv("hikes_fixed.csv", index_col = 0)
```

```python
hikes_df.columns
```

```python
hikes_df_no_na = hikes_df[["Length", "gain", "highpoint", "rating"]].dropna().reset_index(drop = True)
```

```python
hikes_df_no_na.info()
```

```python
X = hikes_df_no_na[["Length", "gain", "highpoint"]]
Y = hikes_df_no_na["rating"]
```

```python
sns.pairplot(hikes_df_no_na)
```

```python
mod1 = sm.OLS(Y, X)
res1 = mod1.fit()
res1.summary()
```

```python
fig = plt.scatter(res1.predict(), res1.resid)
fig.axes.set_ylabel('Residual')
fig.axes.set_xlabel('Prediction') 
```

The R^2 value is surprisingly high for these variables. Judging by the pairplots, they seem to have a loose linear relationship, especially highpoint. It is also interesting how highpoint and Length have a positive signed coefficient, but gain has a negative. Maybe implying that people do not want to work hard to get to a beautiful place. The residuals are also interesting, this is not evenly distributed, it looks almost like the residuals of count data... maybe implying that the number of ratings is more important than any factor.

```python
#hikes_df_no_na["log_gain"] = np.log(hikes_df_no_na["gain"])
X2 = sm.add_constant(hikes_df_no_na[["Length", "log_gain", "highpoint"]])
```

```python
mod2 = sm.OLS(Y, X2)
res2 = mod2.fit()
res2.summary()
```

```python
fig = plt.scatter(res2.predict(), res2.resid)
fig.axes.set_ylabel('Residual')
fig.axes.set_xlabel('Prediction') 
```

This is interesting as the constant is added, the whole model is thrown off, having almost no explantation from the variables. The residuals are more evenly distributed here if you remove the outliers, but none of the variables seem to have any affect, besides the coefficient (which is probably just the average rating, which is able to better predict the rating than the model itself).

```python
from nltk import corpus
from nltk.corpus import stopwords
from nltk import WordNetLemmatizer as wnl
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import time
import re
import glob
```

```python
def text_cleaner(word):
    word = word.lower()
    word = re.sub(r"\'s", ' ', word)
    word = re.sub("<[^>]*>", " ", word)
    word = re.sub('\W', ' ', word)
    word = re.sub('\s+', ' ', word)
    word = word.strip(' ')
    return word
```

```python
hikes_df["summary"] = hikes_df["summary"].fillna("")
```

```python
hikes_df["summary"] = hikes_df["summary"].map(lambda word : text_cleaner(word))
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
x = [None]*len(hikes_df["summary"])
for i in range(len(hikes_df["summary"])):
     x[i] = sid.polarity_scores(hikes_df.iloc[i]["summary"])
```

```python
hikes_df["summary_score"] = [d['compound'] for d in x]
hikes_df['neg'] = [d['neg'] for d in x]
hikes_df['neu'] = [d['neu'] for d in x]
hikes_df['pos'] = [d['pos'] for d in x]
```

```python
hikes_df["summary_score"]
```

```python
hikes_df_no_na = hikes_df[["rating", "Length", "gain", "highpoint", 'Good for kids', 'Fall foliage',
       'Rivers', 'Wildflowers/Meadows', 'Ridges/passes',
       'Dogs allowed on leash', 'Mountain views', 'Old growth', 'Summits',
       'Waterfalls', 'Lakes', 'Established campsites', 'Dogs not allowed',
       'Wildlife', 'Coast', "summary_score", "neg", "neu", "pos"]].dropna().reset_index(drop = True)
X5 = sm.add_constant(hikes_df_no_na[["gain","highpoint","Good for kids", "Dogs allowed on leash", "Mountain views",
                                     "Summits", "Waterfalls", "Dogs not allowed", "Coast", "summary_score", "neu"]])
Y = hikes_df_no_na["rating"]
```

```python
mod5 = sm.OLS(Y, X5)
res5 = mod5.fit()
res5.summary()
```

```python
fig = plt.scatter(res5.predict(), res5.resid)
fig.axes.set_ylabel('Residual')
fig.axes.set_xlabel('Prediction') 
```

Summary score was a good predictor, but probably because it is so average, neg was a terrible predictor, positive sentiment was also bad. Neutral was a better predictor, but just because of its neutrality. These increase the model a small amount, but really offer almost nothing.

## Attmepting to use a Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X, X_test, y, y_test = train_test_split(hikes_df_no_na[["gain","highpoint","Good for kids", "Dogs allowed on leash", "Mountain views",
                                     "Summits", "Waterfalls", "Dogs not allowed", "Coast", "summary_score", "neu"]], 
                                        hikes_df_no_na["rating"], test_size=0.2, random_state=0)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
rfreg = RandomForestRegressor()
rfreg.fit(X_train, y_train)
rfreg.score(X_train,y_train)
```

```python
rfreg.score(X_val, y_val)
```

```python
rfreg = RandomForestRegressor(max_depth=5)
rfreg.fit(X,y)
rfreg.score(X_test, y_test)
```

```python
rfreg.decision_path(X)
```

```python
rfreg.feature_importances_
```
