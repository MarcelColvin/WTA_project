### Adding in VADER Sentiment Analysis

```python
from nltk import corpus
from nltk.corpus import stopwords
from nltk import WordNetLemmatizer as wnl
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import time
import re
import glob
```

```python
def text_cleaner(word):
    word = word.lower()
    word = re.sub(r"\'s", ' ', word)
    word = re.sub("<[^>]*>", " ", word)
    word = re.sub('\W', ' ', word)
    word = re.sub('\s+', ' ', word)
    word = word.strip(' ')
    return word
```

```python
hikes_df["summary"] = hikes_df["summary"].fillna("")
```

```python
hikes_df["summary"] = hikes_df["summary"].map(lambda word : text_cleaner(word))
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
x = [None]*len(hikes_df["summary"])
for i in range(len(hikes_df["summary"])):
     x[i] = sid.polarity_scores(hikes_df.iloc[i]["summary"])
```

```python
hikes_df["summary_score"] = [d['compound'] for d in x]
hikes_df['neg'] = [d['neg'] for d in x]
hikes_df['neu'] = [d['neu'] for d in x]
hikes_df['pos'] = [d['pos'] for d in x]
```

```python
hikes_df["summary_score"]
```

Summary score was a good predictor, but probably because it is so average, neg was a terrible predictor, positive sentiment was also bad. Neutral was a better predictor, but just because of its neutrality. These increase the model a small amount, but really offer almost nothing.

### Grid Search

```python
X_train, X_test, y_train, y_test = train_test_split(hikes_df_rating[["Length", "gain","highpoint", "summary_score", "neu",
                                                        "Lat", "Lon","pos", "neg", "Summits", "Waterfalls", "Mountain views"]], 
                                        hikes_df_rating["rating"], test_size=0.2, random_state=1)
```

```python
rf_model = RandomForestRegressor()
params={"max_depth": range(5,20,5), "min_samples_leaf": range(1,15,5),
         "n_estimators":[200,400]}
model_names =  "rf_model"
parameters_list=[params]
model_log= "_rf"
scaler = StandardScaler()
Grid=GridSearchCV(estimator=rf_model, param_grid=parameters_list, scoring='r2',
                      n_jobs=-1, cv=3, verbose=3)
    
model=Grid.fit(X_train,y_train)
        
globals()['Grid%s' % model_log] =pd.DataFrame(Grid.cv_results_)
```

```python
model.best_score_
```

```python
final_model= model.best_estimator_
```

```python
final_model
```

```python
rfreg4 = final_model
rfreg4.fit(X_train, y_train)
rfreg4.score(X_train,y_train)
```

```python
rfreg4.score(X_val, y_val)
```
